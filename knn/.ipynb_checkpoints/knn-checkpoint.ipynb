kNN(k-Nearest Neighbor)算法是一种常用的监督python dict iteritems学习算法。
给定测试样本，基于某种距离度量找出训练集中与其最靠近的k个训练样本，让后基于这k个邻居的信息进行预测。

预测方式：
- 在分类任务中通常采用“投票法”，即选择这k个样本中出现最多的类别标记作为预测结果；
- 在回归任务中使用“平均法”，即将k个样本的实值输出标记的平均值作为预测结果；
- 还可基于距离远近进行加权平均或加权投票，距离越近的样本权重越大。

1. np.sum 函数

axis表示维度，当数组为二维时，axis=0表示按照x，即按列向加
axis=1表示，按行相加

```
>>> np.sum([[0, 1], [0, 5]], axis=0)
array([0, 6])
>>> np.sum([[0, 1], [0, 5]], axis=1)
array([1, 5])
```
两点假设为(0,0),(1,2)
欧氏距离公式 $
d=\sqrt{(0-1)^2+(0-2)^2}
$


2. argsort函数返回的是数组值从小到大的索引值

```
Examples

One dimensional array:一维数组

>>> x = np.array([3, 1, 2])
>>> np.argsort(x)  
array([1, 2, 0]) 
```

````python
'''
inX 测试集
dataSet 训练集
labels 标签向量
k 邻居的个数
'''
def classify0(inX, dataSet, labels, k):
    # 获取训练集共有多少行
    dataSetSize = dataSet.shape[0]
    # 将目标向量复制与训练集的行数相同，计算欧氏距离
    diffMat = np.tile(inX, (dataSetSize,1)) - dataSet
    sqDiffMat = diffMat**2
    sqDisstance = sqDiffMat.sum(axis=1)
    distance = sqDisstance**0.5
    
    # ,
    classCount = {}
    sortedDistanceIndex = distance.argsort()
    for i in range(k):
        votelabel = labels[sortedDistanceIndex[i]]
        print(votelabel)
        classCount[votelabel] = classCount.get(votelabel,0)+1

    sortedClassCount = sorted(classCount.items(),
                             key=operator.itemgetter(1), reverse=True)
    return sortedClassCount[0][0]
````

## 2.1. 归一化

1. Min-Max Normalization
  
$
x' = \frac{ (x - X_{min})} { (X_{max} - X_{min})}
$


2. 平均归一化
  
$x' = \frac{(x - μ)} {(MaxValue - MinValue)}$

（1）和（2）有一个缺陷就是当有新数据加入时，可能导致max和min的变化，需要重新定义。

3. 非线性归一化
  
1）对数函数转换：y = log10(x)  
2）反余切函数转换：y = atan(x) * 2 / π
   
非线性经常用在数据分化比较大的场景，有些数值很大，有些很小。通过一些数学函数，将原始值进行映射。该方法包括 log、指数，正切等。需要根据数据分布的情况，决定非线性函数的曲线，比如log(V, 2)还是log(V, 10)等。


````python
# datingDataSet.max()

def normalize(dataSet):
    maxnum = dataSet.max()
    minnum = dataSet.min()
    ranges = maxnum - minnum
    result = (dataSet - minnum)/(maxnum-minnum)
    
    return ranges, result

ranges, result = normalize(datingDataSet)
````

````python
# 划分数据集
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(result, datingLabels, test_size=0.33, random_state=42)
````

````python
def classify0(inX, dataSet, labels, k):
    # 获取训练集共有多少行
    dataSetSize = dataSet.shape[0]
    # 将目标向量复制与训练集的行数相同，计算欧氏距离
    diffMat = np.tile(inX, (dataSetSize,1)) - dataSet
    sqDiffMat = diffMat**2
    sqDisstance = sqDiffMat.sum(axis=1)
    distance = sqDisstance**0.5
    
    classCount = {}
    sortedDistanceIndex = distance.argsort()
    for i in range(k):
        votelabel = labels[sortedDistanceIndex[i]]
        classCount[votelabel] = classCount.get(votelabel,0)+1

    sortedClassCount = sorted(classCount.items(),
                             key=operator.itemgetter(1), reverse=True)
    return sortedClassCount[0][0]
````

````python
error = 0

X_test_arr = np.array(X_test)
y_test = np.array(y_test)

for i in range(len(X_test_arr)):
    X_train = pd.DataFrame(X_train)[[0,1,2]]
    y_train = pd.DataFrame(y_train)[0]
    classfied = classify0(X_test_arr[i], X_train, y_train, 10)
    print("分类值为 %s, 实际分类为 %s" %(classfied, y_test[i]))
    
    if classfied!=y_test[i]:
        error += 1
print("error = %f " % (error/float(len(X_test_arr))))
````
